```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.shape
```

```python
df.drop('id', axis = 1, inplace= True)
```

```python
df.drop('host_id', axis = 1, inplace= True)
```

```python
df.describe()
```

```python
df['price_per_night'] = df.price / df.minimum_nights
df.head()
```

```python
df.describe()
```

```python
min_t, max_t = df.price_per_night.quantile([0.01, 0.99])
min_t, max_t
```

```python
df.drop(['name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'number_of_reviews'], axis = 1, inplace= True)
df['last_review'] = pd.to_datetime(df['last_review']).astype(int)
df.reviews_per_month.fillna(0, inplace= True)
```

```python
room_type_dummies = pd.get_dummies(df['room_type'], prefix='room_type', drop_first=True)
```

```python
df = pd.concat([df, room_type_dummies], axis=1)
```

```python
df.drop('room_type', axis = 1, inplace = True)
```

```python
df[df['price_per_night'] > max_t]
```

```python
df[df['price_per_night'] < min_t]
```

```python
df['latitude'] = df['latitude'].round().astype(int)
df['longitude'] = df['longitude'].round().astype(int)
df['price_per_night'] = df['price_per_night'].round().astype(int)
df['reviews_per_month'] = df['reviews_per_month'].round().astype(int)
```

```python
df_no_out = df[(df['price_per_night'] > min_t)&(df['price_per_night'] < max_t)]
df_no_out.shape
```

```python
df_no_out.sample(5)
```

```python
df_no_out.head()
```

```python
df_no_out.isna().sum().values
```

```python
X = df_no_out.drop('price_per_night', axis = 1)
y = df_no_out.price_per_night
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state = 30)
```

```python
y.shape
```

```python
X.shape
```

```python
randforest = RandomForestClassifier()
randforest.fit(X_train, y_train)
```

```python
randforest.score(X_test, y_test)
```
